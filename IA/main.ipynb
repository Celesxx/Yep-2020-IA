```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import rcParams

import tensorflow as tf
import numpy as np
import pandas as pd
import sys
import cv2
```

## Loading d'une image

```python
img_box =  cv2.cvtColor(cv2.imread('../image/blue.jpg'), cv2.COLOR_BGR2RGB)
img_bottle = cv2.cvtColor(cv2.imread('../image/bottle.jpg'), cv2.COLOR_BGR2RGB)
img_bouchon = cv2.cvtColor(cv2.imread('../image/bouchon.jpg'), cv2.COLOR_BGR2RGB)
img_pince = cv2.cvtColor(cv2.imread('../image/pince.jpg'), cv2.COLOR_BGR2RGB)
img_list = [img_box, img_bottle, img_bouchon, img_pince]


%matplotlib inline

# figure size in inches optional
rcParams['figure.figsize'] = 11 ,8

# display images
fig, ax = plt.subplots(2,2)
ax[0,0].imshow(img_list[0])
ax[0,1].imshow(img_list[1])
ax[1,0].imshow(img_list[2])
ax[1,1].imshow(img_list[3])
```

```python
result = []
for img in img_list:
    blurred_frame = cv2.GaussianBlur(img, (5, 5), 0)
    hsv = cv2.cvtColor(blurred_frame, cv2.COLOR_RGB2HSV)

    lower_blue = np.array([38, 86, 0])
    upper_blue = np.array([121, 255, 255])
    mask = cv2.inRange(hsv, lower_blue, upper_blue)
    contours, hierarchy = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    result.append(cv2.drawContours(img, contours, -1, (0, 255, 0), 3))


fig, ax = plt.subplots(2,2)
ax[0,0].imshow(result[0])
ax[0,1].imshow(result[1])
ax[1,0].imshow(result[2])
ax[1,1].imshow(result[3])
```

```python
img_box1 =  cv2.cvtColor(cv2.imread('../image/blue.jpg'), cv2.COLOR_BGR2GRAY)
img_bottle1 = cv2.cvtColor(cv2.imread('../image/bottle.jpg'), cv2.COLOR_BGR2GRAY)
img_bouchon1 = cv2.cvtColor(cv2.imread('../image/bouchon.jpg'), cv2.COLOR_BGR2GRAY)
img_pince1 = cv2.cvtColor(cv2.imread('../image/pince.jpg'), cv2.COLOR_BGR2GRAY)
img_list1 = [img_box1, img_bottle1, img_bouchon1, img_pince1]

fig, ax = plt.subplots(2,2)
ax[0,0].imshow(img_list1[0])
ax[0,1].imshow(img_list1[1])
ax[1,0].imshow(img_list1[2])
ax[1,1].imshow(img_list1[3])
```

```python
result = []
i = 0
for img in img_list1:
    _,mask1 = cv2.threshold(img,100,255,cv2.THRESH_BINARY)
    contours, hierarchy = cv2.findContours(mask1, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    result.append(cv2.drawContours(img_list[i], contours, -1, (0, 255, 0), 3))
    i += 1

fig, ax = plt.subplots(2,2)
ax[0,0].imshow(result[0])
ax[0,1].imshow(result[1])
ax[1,0].imshow(result[2])
ax[1,1].imshow(result[3])
```
